# Applying different Algorithms

```python
train_data = df.drop(['Year','Country',"Status"],axis=1)
train_data.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = train_data.drop('Life expectancy ',axis=1)
y = train_data['Life expectancy ']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression 
regr = LinearRegression() 
regr.fit(X_train, y_train) 
```

```python
print(regr.score(X_test, y_test)) 
```

```python
from sklearn.ensemble import RandomForestRegressor 
  
 # create regressor object 
regressor = RandomForestRegressor(n_estimators = 100,max_depth=20, random_state = 2) 
  
# fit the regressor with x and y data 
regressor.fit(X_train, y_train) 
```

```python
regressor.score(X_test,y_test)
```

# Saving model using joblib

```python

from sklearn.externals import joblib 
  
# Save the model as a pickle in a file 
joblib.dump(regressor, 'predict_life.pkl') 
  
# Load the model from the file 
knn_from_joblib = joblib.load('predict_life.pkl')  
  
# Use the loaded model to make predictions 
knn_from_joblib.predict(X_test) 
```
